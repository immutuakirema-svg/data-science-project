```python
import numpy as np
import matplotlib.pyplot as plt

yaxis= np.array([40,30,10,50,25,5])
xaxis=('crime','science fiction','drama','comedy','action','documentary')
plt.figure(figsize=(8,6))
plt.title('JIM\'s VIDEO LIBRARY', fontsize=20, color='blue', fontweight='bold')
plt.xlabel('GENRES', fontsize=15, color='red', fontweight='bold')
plt.ylabel('NUMBER OF MOVIES', fontsize=15, color='red', fontweight='bold')
plt.bar(xaxis,yaxis, color='black')
plt.show()
```

```python
# Replace None with appropriate code

weight = [2750, 3125, 2100, 4082, 2690, 3640, 4380, 2241, 2895, 3659]
mpg = [29, 23, 33, 28, 20, 21, 14, 25, 31, 17]

# Create the plot
plt.figure(figsize=(8,6))

# Plot with scatter()
plt.scatter(weight, mpg, color='blue', marker='o')

# Set x and y axes labels, legend, and title
xaxis = 'Weight (lbs)'
yaxis = 'Miles Per Gallon (MPG)'
plt.xlabel(xaxis, fontsize=15, color='red', fontweight='bold')
plt.ylabel(yaxis, fontsize=15, color='red', fontweight='bold')
plt.title('CONSUMER CARS', fontsize=20, color='blue', fontweight='bold')
plt.legend(['MPG vs Weight'], loc='upper right')
plt.show()
```

```python
# Replace None with appropriate code

x = [43.1, 35.6, 37.5, 36.5, 45.3, 43.4, 
     40.3, 50.2, 47.3, 31.2, 42.2, 45.5, 
     30.3, 31.4, 35.6, 45.2, 54.1, 45.6, 
     36.5, 43.1]

# Create the plot
plt.figure(figsize=(8,6))

# Plot the histogram with hist() function
plt.hist(x, bins=5, color='green')

# Label axes and set title
plt.xlabel('Waiting Times (minutes)', fontsize=15, color='red', fontweight='bold')
plt.ylabel('Number of Customers', fontsize=15, color='red', fontweight='bold')
plt.title('CUSTOMER WAITING TIMES', fontsize=20, color='blue', fontweight='bold')
plt.show()
```
